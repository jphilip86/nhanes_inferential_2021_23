```python
# 2.1 Marital Status → married (binary) 2.1 Marital Status (DMDMARTZ)/ CATEGORICAL
# Codebook values:
# 1 = Married/Living with partner (Count: 4136)
# 2 = Widowed/Divorced/Separated (Count: 2022)
# 3 = Never married (Count: 1625)
# 77 = Refused (Count: 4)
# 99 = Don't know (Count: 5)
# Missing (Count: 4141)

def recode_marital(x):
    """
    NHANES DMDMARTZ codes:
    1 = Married/Living with partner → 1
    2 = Widowed/Divorced/Separated → 0
    3 = Never married → 0
    77, 99, NaN = Refused/Don't know/Missing → NaN
    """
    if x == 1:
        return 1        #  If their answer is 1 (Married/Living together), we call it 1
    elif x in [2, 3]:
        return 0        #  If their answer is 2 or 3 (Not married), we call it 0 (no)
    elif x in [77, 99] or pd.isna(x):
        return np.nan   # If their answer is 77 or 99 (Refused or Don’t know), or they didn’t answer,
    else:
        return np.nan   # If for some reason the code is different, we also call it missing (blank)

# Apply the recoding to the 'DMDMARTZ' column

demo['married'] = demo['DMDMARTZ'].apply(recode_marital)
print("Marital Status Recoding:")
print(demo['married'].value_counts(dropna=False))
```

```python
# 2.2 Education Level → bachelor_or_higher 2.2 Education Level (DMDEDUC2) / CATEGORICAL
# Codebook values:
# 1 = Less than 9th grade (Count: 373)
# 2 = 9-11th grade (Count: 666)
# 3 = High school graduate/GED (Count: 1749)
# 4 = Some college or AA degree (Count: 2370)
# 5 = College graduate or above (Count: 2625)
# 7 = Refused (Count: 0)
# 9 = Don't know (Count: 11)
# Missing (Count: 4139)

def recode_education(x):
    """
    5 = College graduate or above → 1 (bachelor's+)
    1-4 = Less than bachelor's → 0
    7, 9, NaN = Unknown → NaN
    """
    if x == 5:
        return 1
    elif x in [1, 2, 3, 4]:
        return 0
    elif x in [7, 9] or pd.isna(x):
        return np.nan
    else:
        return np.nan

# Apply the recoding to the 'DMDEDUC2' column
demo['bachelor_or_higher'] = demo['DMDEDUC2'].apply(recode_education)
print("\nEducation Level Recoding:")
print(demo['bachelor_or_higher'].value_counts(dropna=False))
```

```python
# 2.3 Kidney Health (KIQ022)  / CATEGORICAL
# Codebook values:
# 1 = Yes (Count: 321)
# 2 = No (Count: 7473)
# 7 = Refused (Count: 0)
# 9 = Don't know (Count: 13)
# Missing (Count: 2)

def recode_kidney(x):
    """
    1 = Yes, told had weak/failing kidneys → 1
    2 = No → 0
    7, 9, NaN = Unknown → NaN
    """
    if x == 1:
        return 1
    elif x == 2:
        return 0
    elif x in [7, 9] or pd.isna(x):
        return np.nan
    else:
        return np.nan

# Apply the recoding to the 'KIQ022' column
kiq['weak_kidney'] = kiq['KIQ022'].apply(recode_kidney)
print("\nKidney Health Recoding:")
print(kiq['weak_kidney'].value_counts(dropna=False))
```

```python
# 2.4 Sedentary Minutes (PAD680)Codebook values:  /CONTINUOUS
# 0 to 1380 = Range of Values (Count: 8065)
# 7777 = Refused (Count: 6)
# 9999 = Don't know (Count: 67)
# Missing (Count: 15)
# Replace 7777, 9999 with NaN


paq['PAD680'] = paq['PAD680'].replace([7777, 9999], np.nan)
print("\nSedentary Minutes Cleaning:")
print(f"Valid values: {paq['PAD680'].notna().sum()}")
print(f"Missing: {paq['PAD680'].isna().sum()}")
```

```python
# 2.5 Self-Reported Weight (WHD020)  /CONTINUOUS
# Codebook values:
# 63 to 530 = Range of Values (Count: 8358)
# 7777 = Refused (Count: 40)
# 9999 = Don't know (Count: 88)
# Missing (Count: 15)
# Replace 7777, 9999 with NaN

whq['WHD020'] = whq['WHD020'].replace([7777, 9999], np.nan)
print("\nWeight Cleaning:")
print(f"Valid values: {whq['WHD020'].notna().sum()}")
print(f"Missing: {whq['WHD020'].isna().sum()}")
```

```python
# 2.6 Vitamin D Lab (LBDVD2LC)  / CATEGORICAL
# Codebook values:
# 0 = At or above the detection limit (Count: 1447)
# 1 = Below lower detection limit (Count: 5860)
# Missing (Count: 1420)
# Keep as-is, it's already binary

def recode_vitamin_d(x):
    """
    Recodes LBDVD2LC (Vitamin D status):
    0 = Adequate (at or above detection limit)
    1 = Low (below lower detection limit)
    . = NaN (missing)
    """
    if x == 0:
        return 0    # Adequate
    elif x == 1:
        return 1    # Low
    else:
        return np.nan   # Missing or any unusual value

# Apply the recoding to the 'LBDVD2LC' column
vid['vitamin_d_status'] = vid['LBDVD2LC'].apply(recode_vitamin_d)

print("\nVitamin D Lab Status:")
print(vid['vitamin_d_status'].value_counts(dropna=False))
```

```python
# 2.7 Hepatitis B Surface Antibody (LBXHBS)  / CATEGORICAL
# Codebook values:
# 1 = Positive (Count: 2042)
# 2 = Negative (Count: 5324)
# 3 = Indeterminate (Count: 0)
# Missing (Count: 1245)


def recode_hepb(x):
    """
    1 = Positive → 1
    2 = Negative → 0
    3, NaN = Indeterminate/Missing → NaN
    """
    if x == 1:
        return 1
    elif x == 2:
        return 0
    elif x == 3 or pd.isna(x):
        return np.nan
    else:
        return np.nan

# Apply the recoding to the 'LBXHBS' column
hepb['hepb_positive'] = hepb['LBXHBS'].apply(recode_hepb)
print("\nHepatitis B Antibody Recoding:")
print(hepb['hepb_positive'].value_counts(dropna=False))
```

```python
# 2.8 Blood Pressure (BPXOSY3, BPXODI3)  /CONTINUOUS
# Codebook values:
# BPXOSY3: 50 to 232 mmHg (Count: 7480), Missing (321)
# BPXODI3: 24 to 136 mmHg (Count: 7480), Missing (321)
# No cleaning needed - continuous values are valid

pd.options.display.float_format = '{:.2f}'.format  # Show decimals, not scientific notation

# Filter only valid systolic readings: 50-232 mmHg
sys_valid = bpxo['BPXOSY3'][(bpxo['BPXOSY3'] >= 50) & (bpxo['BPXOSY3'] <= 232)]
print("\nBlood Pressure - Systolic (Valid Range Only, mmHg):")
print(sys_valid.describe())

# Filter only valid diastolic readings: 24-136 mmHg
dia_valid = bpxo['BPXODI3'][(bpxo['BPXODI3'] >= 24) & (bpxo['BPXODI3'] <= 136)]
print("\nBlood Pressure - Diastolic (Valid Range Only, mmHg):")
print(dia_valid.describe())
```

```python
# 2.9 Age (RIDAGEYR) /CONTINUOUS
# Codebook values:
# 0 to 79 = Range of Values (Count: 11408)
# 80 = 80 years of age and over (Count: 525)
# Missing (Count: 0)
# No recoding needed - age is continuous


pd.options.display.float_format = '{:.2f}'.format  # Show floats as fixed point

# Filter: Only accepted ages (NHANES: 0 to 80; 80 means "80 or older", no missing)
age_valid = demo['RIDAGEYR'][(demo['RIDAGEYR'] >= 0) & (demo['RIDAGEYR'] <= 80)]

print("\nAge Distribution (Valid NHANES Codes Only):")
print(age_valid.describe())
```

```python
# Step 3: Merge All DataFrames by SEQN

# Start with demographics
merged = demo[['SEQN', 'RIDAGEYR', 'RIAGENDR', 'married', 'bachelor_or_higher']].copy()

# Merge kidney
merged = merged.merge(kiq[['SEQN', 'weak_kidney']], on='SEQN', how='left')

# Merge blood pressure
merged = merged.merge(bpxo[['SEQN', 'BPXOSY3', 'BPXODI3']], on='SEQN', how='left')

# Merge physical activity
merged = merged.merge(paq[['SEQN', 'PAD680']], on='SEQN', how='left')

# Merge weight
merged = merged.merge(whq[['SEQN', 'WHD020']], on='SEQN', how='left')

# Merge vitamin D
merged = merged.merge(vid[['SEQN', 'vitamin_d_status']], on='SEQN', how='left')

# Merge hepatitis B
merged = merged.merge(hepb[['SEQN', 'hepb_positive']], on='SEQN', how='left')

print(f"\n✓ Final merged dataset shape: {merged.shape}")
display(merged.head())
```

```python
# Inferential Statistics:

# Question 1: Association Between Marital Status and Education Level
# Research Question: Is there an association between marital status (married or not married) and education level (bachelor's degree or higher vs. less than a bachelor's degree)?
# Variables: married (recoded from DMDMARTZ) and bachelor_or_higher (recoded from DMDEDUC2)
# Appropriate Test: Chi-Square Test of Independence (both variables are categorical)


print("="*70) # creates a visual line by repeating the “=” symbol 70 times
print("QUESTION 1: Association Between Marital Status and Education Level")
print("="*70)

# Remove missing values
q1_data = merged[['married', 'bachelor_or_higher']].dropna() # Selects only the two relevant categorical variables.

# Create contingency table
crosstab = pd.crosstab(q1_data['married'], q1_data['bachelor_or_higher'], margins=True)
print("\nContingency Table (Marital Status vs Education):")
print(crosstab)

# Perform chi-square test # compares the observed counts in the crosstab to expected counts (assuming no association).
# chi2: the Chi-Square statistic — measures deviation of observed from expected counts.
# the p-value — probability of seeing this result by chance.
# dof: degrees of freedom

chi2, p, dof, expected = stats.chi2_contingency(
    pd.crosstab(q1_data['married'], q1_data['bachelor_or_higher'])
)

print(f"\nChi-Square Test Results:")
print(f"  Chi-square statistic: {chi2:.3f}")
print(f"  p-value: {p:.4f}")
print(f"  Degrees of freedom: {dof}")
print(f"  Sample size: {len(q1_data)}")

# Interpretation
if p < 0.05:
    print(f"\n✓ SIGNIFICANT: There IS a statistically significant association")
    print(f"  between marital status and education level (p < 0.05)")
else:
    print(f"\n✗ NOT SIGNIFICANT: There is NO statistically significant association")
    print(f"  between marital status and education level (p ≥ 0.05)")


```

```python
# 5.2 Question 2: T-Test - Sedentary Behavior by Marital Status
# Research Question: Do married individuals differ from not married individuals in their average daily sedentary time (in minutes)?

print("\n" + "="*70)
print("QUESTION 2: Sedentary Behavior by Marital Status")
print("="*70)

# Remove missing values
# Selects only the columns needed (married and PAD680, which holds sedentary time in minutes).
q2_data = merged[['married', 'PAD680']].dropna()

# Separate groups
married_group = q2_data[q2_data['married'] == 1]['PAD680']
not_married_group = q2_data[q2_data['married'] == 0]['PAD680']

# Descriptive statistics # Calculates group size (len()), mean (.mean()), and standard deviation (.std()).
print(f"\nDescriptive Statistics:")
print(f"  Married: n={len(married_group)}, Mean={married_group.mean():.2f} min, SD={married_group.std():.2f}")
print(f"  Not Married: n={len(not_married_group)}, Mean={not_married_group.mean():.2f} min, SD={not_married_group.std():.2f}")
print(f"  Mean difference: {married_group.mean() - not_married_group.mean():.2f} minutes")

# Independent samples t-test # tests whether the two independent groups have equal means
# t-statistic: measures how far apart the means are relative to variability.
# p-value: probability that such difference occurred by chance if both groups’ means were actually equal.

t_stat, p_val = stats.ttest_ind(married_group, not_married_group)

print(f"\nIndependent T-Test Results:")
print(f"  t-statistic: {t_stat:.3f}")
print(f"  p-value: {p_val:.4f}")

# Interpretation
if p_val < 0.05:
    print(f"\n✓ SIGNIFICANT: There IS a statistically significant difference in")
    print(f"  sedentary time between married and not married groups (p < 0.05)")
else:
    print(f"\n✗ NOT SIGNIFICANT: There is NO statistically significant difference in")
    print(f"  sedentary time between married and not married groups (p ≥ 0.05)")
```

```python
# 5.3 Question 3: Multiple Regression - Age and Marital Status Effect on Systolic BP
# Appropriate Test: Multiple Linear Regression (predicting continuous outcome from continuous and categorical predictors)

# Section Header
print("\n" + "="*70)
print("QUESTION 3: Age and Marital Status Effects on Systolic BP")
print("="*70)

# Remove missing values
# Selecting three columns from the dataset:
# RIDAGEYR: participant’s age in years (continuous variable).
# married: coded marital status (1 = married, 0 = not married).
# BPXOSY3: systolic blood pressure measurement (in mmHg).
# .dropna() removes any row that contains missing data in these columns—essential for valid regression
# analysis (missing values break model fitting).

q3_data = merged[['RIDAGEYR', 'married', 'BPXOSY3']].dropna()

# Prepare data for regression Defines independent (predictor) variables X — age and marital status.
# Defines dependent (target) variable y — systolic blood pressure.
X = q3_data[['RIDAGEYR', 'married']]
y = q3_data['BPXOSY3']

# Fit linear regression model
model = LinearRegression() # creates a scikit-learn regression model.
model.fit(X, y) # calculates the best-fit line coefficients by minimizing the sum of squared residuals (Least Squares Method).
y_pred = model.predict(X) # uses that fitted model to forecast systolic BP for every observation (producing the model’s predicted values

# Results
print(f"\nMultiple Linear Regression Results:")
print(f"  Sample size: {len(q3_data)}") # Displays sample size (number of participants) using len(q3_data)
print(f"  R-squared: {r2_score(y, y_pred):.4f}") # It tells how much of the variation in blood pressure is explained by age and marital status combined.
# Values close to 1 indicate a strong fit.

# Coefficient Interpretation
print(f"\nRegression Coefficients:")
print(f"  Intercept: {model.intercept_:.3f} mmHg") # Intercept: Predicted BP when both predictors (age and marital status)
print(f"  Age coefficient: {model.coef_[0]:.3f} mmHg per year") # Age coefficient: Predicted change in systolic BP for each 1-year increase in age.
print(f"  Marital status coefficient: {model.coef_[1]:.3f} mmHg") # Marital coefficient: Average BP difference between married (1) and not married (0) individuals, controlling for age

# Individual correlations
r_age, p_age = stats.pearsonr(q3_data['RIDAGEYR'], q3_data['BPXOSY3']) # calculates Pearson’s correlation coefficient (r) and its p-value:
print(f"\nAge-BP Correlation:")
print(f"  Pearson r: {r_age:.3f}, p-value: {p_age:.4f}") # r shows linear strength and direction

# Marital status effect (t-test) # Splits the dataset into two groups based on marital status
# Runs independent-samples t-test (ttest_ind) comparing group means
# t_stat = standardized difference of means.
# p_marital = probability of getting such a difference by random chance.​

married_bp = q3_data[q3_data['married'] == 1]['BPXOSY3']
not_married_bp = q3_data[q3_data['married'] == 0]['BPXOSY3']
t_stat, p_marital = stats.ttest_ind(married_bp, not_married_bp)

# Reporting T-Test Results
print(f"\nMarital Status Effect on BP:") # Prints mean systolic BP and sample count per group.
print(f"  Married mean: {married_bp.mean():.2f} mmHg (n={len(married_bp)})")
print(f"  Not married mean: {not_married_bp.mean():.2f} mmHg (n={len(not_married_bp)})")
print(f"  t-statistic: {t_stat:.3f}, p-value: {p_marital:.4f}") # Displays the t-statistic and p-value from the test above.


# Interpretation
print(f"\n INTERPRETATION:")
if p_age < 0.05:
    print(f"  ✓ Age significantly affects systolic BP (p < 0.05)")
else:
    print(f"  ✗ Age does NOT significantly affect systolic BP (p ≥ 0.05)")

if p_marital < 0.05:
    print(f"  ✓ Marital status significantly affects systolic BP (p < 0.05)")
else:
    print(f"  ✗ Marital status does NOT significantly affect systolic BP (p ≥ 0.05)")
```

```python
# 5.4 Question 4: Pearson Correlation - Weight vs Sedentary Behavior
# This code measures how weight (WHD020) and sedentary behavior (PAD680)
# are linearly related using Pearson correlation

print("\n" + "="*70) # Section Header
print("QUESTION 4: Correlation Between Weight and Sedentary Behavior")
print("="*70)

# Remove missing values


q4_data = merged[['WHD020', 'PAD680']].dropna()

# Calculate Pearson correlation
r, p_val = stats.pearsonr(q4_data['WHD020'], q4_data['PAD680'])

print(f"\nPearson Correlation Results:")
print(f"  Sample size: {len(q4_data)}") #Total number of records used after removing missing values.
print(f"  Pearson r: {r:.3f}") # Displays the correlation strength and direction (rounded to three decimals).
print(f"  p-value: {p_val:.4f}") # Tests statistical significance; a smaller p (< 0.05) implies the relationship is unlikely due to random chance.
print(f"  R-squared: {r**2:.4f} ({r**2*100:.2f}% of variance explained)") # proportion of variance in sedentary behavior explained by weight (square of r)

# Descriptive statistics
print(f"\nDescriptive Statistics:") # Prints mean (average) and standard deviation (spread of data) for each variable.
print(f"  Weight: Mean={q4_data['WHD020'].mean():.2f} lbs, SD={q4_data['WHD020'].std():.2f}") # Mean indicates the central tendency.
print(f"  Sedentary time: Mean={q4_data['PAD680'].mean():.2f} min, SD={q4_data['PAD680'].std():.2f}")

# Interpretation
print(f"\n INTERPRETATION:")
if p_val < 0.05:
    if abs(r) < 0.3:
        strength = "weak"
    elif abs(r) < 0.7:
        strength = "moderate"
    else:
        strength = "strong"

    direction = "positive" if r > 0 else "negative"
    # r > 0: positive → as weight increases, sedentary behavior tends to increase.
    # r < 0: negative → as weight increases, sedentary behavior tends to decrease.

    print(f"  ✓ SIGNIFICANT: There IS a statistically significant {strength} {direction}")
    print(f"    correlation between weight and sedentary behavior (p < 0.05)")
else:
    print(f"  ✗ NOT SIGNIFICANT: There is NO statistically significant correlation")
    print(f"    between weight and sedentary behavior (p ≥ 0.05)")
```

```python
# ==========================================
# Q5: CREATIVE ANALYSIS
# Hepatitis B Antibody Status and Blood Pressure
# States the hypothesis: whether Hepatitis B antibody status relates to systolic blood pressure.
# ==========================================

print("\n" + "="*70)
print("QUESTION 5: CREATIVE ANALYSIS")
print("Hepatitis B Antibody Status and Systolic Blood Pressure")
print("="*70)

print("\n RESEARCH QUESTION:")
print("Is there a difference in systolic blood pressure between individuals")
print("with positive hepatitis B antibodies vs. those with negative antibodies?")

print("\n TEST CHOICE: Independent Samples T-Test")
print("RATIONALE: Comparing mean systolic BP between two independent groups")
print("(positive vs. negative hepatitis B antibody status)")

# Appropriate because:Both groups are independent (different subjects).
# Dependent variable (BP) is continuous
# Predictor (antibody status) is binary categorical.

print("\n WHY THIS QUESTION?")
print("Hepatitis B infection and immune response may be associated with")
print("cardiovascular outcomes. This exploratory analysis examines whether")
print("antibody status relates to blood pressure levels.")

# Remove missing values
q5_data = merged[['hepb_positive', 'BPXOSY3']].dropna()

# Data Quality check
print(f"\n DATA CHECK:")
print(f"Total valid observations: {len(q5_data)}")
print(f"\nHepatitis B Antibody Status Distribution:")
print(q5_data['hepb_positive'].value_counts(dropna=False))

# Separate groups # Filters the dataset to include only:
# hepb_positive: 1 for positive antibodies, 0 for negative.
# BPXOSY3: systolic blood pressure

negative_hepb = q5_data[q5_data['hepb_positive'] == 0.0]['BPXOSY3']
positive_hepb = q5_data[q5_data['hepb_positive'] == 1.0]['BPXOSY3']


# Descriptive Statistics # Provides sample counts, mean BP, and standard deviation (measure of BP variability)
print(f"\n DESCRIPTIVE STATISTICS:")
print(f"  Negative HepB Antibodies:")
print(f"    n = {len(negative_hepb)}")
print(f"    Mean = {negative_hepb.mean():.2f} mmHg")
print(f"    SD = {negative_hepb.std():.2f} mmHg")

print(f"\n  Positive HepB Antibodies:")
print(f"    n = {len(positive_hepb)}")
print(f"    Mean = {positive_hepb.mean():.2f} mmHg")
print(f"    SD = {positive_hepb.std():.2f} mmHg")

mean_diff = positive_hepb.mean() - negative_hepb.mean()
print(f"\n  Mean difference: {mean_diff:.2f} mmHg")

# Independent samples t-test
t_stat, p_val = stats.ttest_ind(negative_hepb, positive_hepb)

print(f"\n INDEPENDENT T-TEST RESULTS:")
print(f"  t-statistic: {t_stat:.3f}")
print(f"  p-value: {p_val:.4f}")

# Interpretation Logic
print(f"\n FINDINGS:")
if p_val < 0.05:
    print(f"  ✓ SIGNIFICANT: There IS a statistically significant difference in")
    print(f"    systolic BP between hepatitis B antibody groups (p < 0.05)")
    if positive_hepb.mean() > negative_hepb.mean():
        print(f"    → Individuals with POSITIVE antibodies have HIGHER blood pressure")
        print(f"      ({positive_hepb.mean():.1f} vs {negative_hepb.mean():.1f} mmHg)")
    else:
        print(f"    → Individuals with NEGATIVE antibodies have HIGHER blood pressure")
        print(f"      ({negative_hepb.mean():.1f} vs {positive_hepb.mean():.1f} mmHg)")
else:
    print(f"  ✗ NOT SIGNIFICANT: There is NO statistically significant difference in")
    print(f"    systolic BP between hepatitis B antibody groups (p ≥ 0.05)")
    print(f"    Both groups have similar average blood pressure levels.")

print("\n INTERPRETATION:")
print("This analysis examined whether immune response to hepatitis B")
print("(indicated by antibody presence) is associated with cardiovascular")
print("markers like blood pressure. The results suggest" +
      (" a significant" if p_val < 0.05 else " no significant") +
      " relationship.")


```
